```python
def SET_Node(tree, node_name, left_decision, right_decision, left_feature, right_feature):
  tree.append({node_name: {'left_decision': left_decision, 'right_decision': right_decision, 
                           'left_feature': left_feature, 'right_feature': right_feature}})

def GET_sub_DataSet(data_frame, feature_to_remove, value):
  
  ''' This function returns DF_0 and DF_1 where it gets the feature by 
      value and removes the feature column from the DataSet returned '''

  DataSet = (data_frame.loc[data_frame[feature_to_remove] == value]).drop([feature_to_remove], axis = 1)
  return DataSet, feature_to_remove

def GET_labels_by_value(df_data, feature_name, value):
  ''' This function gets sub set of labels that follow a certain value.
      i.e. for a feature x; it takes binary values either 0 or 1.
      for a certain feature value i.e. 0, get all the labels corresponding
      to this feature value.
      Return: D_Labels -> this is a sub dataset that contains all the labels
      corresponding to a single feature value from the feature values taken by 
      a single feature.'''
  
  # Get all the labels from the dataframe
  labels = df_data['Label'].tolist()
  
  # Get all feature values for the needed feature
  feature_values = df_data[feature_name].tolist()

  # find the location of 'value' in the feature values
  feature_zero_indices  = np.where(np.array(feature_values) == value)[0]

  # Get the sub DataSet; D_Labels for a feature, where it has a single feature value i.e. 0
  D_labels = [labels[i] for i in feature_zero_indices.tolist()]

  return D_labels
```

```python
# The decision tree is saved in the TREE list, where each entry is a whole node
# with its decisions and splitting features. In order to avoid certain errors and
# make life easier, there is a root node. This node is just a trivial node with no 
# purpose except for making the code designed easier for me.
TREE = []
MAX_DEPTH = 0
SET_Node(TREE, 'root', 1, 1, 'x', 'x')

def EntropyAndError_calculation(data, true_data_length, feature_name):

  ''' 
      This function calculate entropy for data of a dataset. 
      The entorpy equation is H(Y) = -SUM { P(Y=y) * log_2(P(Y=y) }.
      It can be used to calculate any entropy value i.e. H(Y), H(Y|A=0),
      H(Y|A=1) or any other H related function.
      
      The funciton counts number of zeros and ones in the argument dataset.
      It then computes P(Y=0) and P(Y=1) for a binary values.  
      Note: This function works only for binary classification
      : return: the entropy value, error value, and the decision for this node
  '''

  # Get number of 1s and 0s for the data sent to this function
  my_dict = {i:data.count(i) for i in data} 
  
  # Handle zero 1s or zero 0s in the DataSet by hardcoding them to value: 0
  if 0 not in my_dict:
    my_dict[0] = 0
  if 1 not in my_dict:
    my_dict[1] = 0


  # This function decides which value to assign this node we calculate its entropy
  # If number of 1s are greater than 0s, then assign this decision to 1. Else 0
  if my_dict[0] < my_dict[1]:
    node_decision = 1
  else:
    node_decision = 0

  # This line is to avoid an error, maybe removed later during code refactoring
  if true_data_length == 0:
    return -10, -10, node_decision
  
  # Calculate probabilty of P(Y=0) and P(Y=1) to calculate the entropy 
  prob_y_0 = my_dict[0] / true_data_length
  prob_y_1 = my_dict[1] / true_data_length

  # Make sure that P(y=0) and P(Y=1) are finite numbers > 0. If they were 0,
  # reset them to 1 to remove them from the equation by using log base 2 for the P(Y=x)
  # Note: for P(Y) = 0, this case is undefined and needs to be neglected, that's why
  # i replaced the 0 probability with 1 probabilty, to get entropy value = 0 and avoid
  # choosing this feature for splitting.
  if prob_y_0 == 0:
    prob_y_0 = 1
  
  if prob_y_1 == 0:
    prob_y_1 = 1
    
  # Simply applies the entropy equation in the desciption
  entropy_value = -prob_y_0 * np.log2(prob_y_0) + -prob_y_1 * np.log2(prob_y_1)

  # Error rate for majority vote classifier -> Count of most common label / len(Total amount of data)
  # i.e. if the total data has 28 examples and we are working on subset of this data with length 14
  # In this case we calculate accuracy: (14 - number of most common value i.e. 10 1s or 20 0s) divide by 28
  error_rate = (len(data) - my_dict[node_decision]) / true_data_length
  
  return entropy_value, error_rate, node_decision

def MutualInformationGain_calculator(df_data, feature_name, decision, total_length_of_whole_DS):

  ''' 
      This function calculats the mutual information gain. It needs the
      whole dataframe as in order to calculate H(Y) and the preceeds with 
      the calculations to get the mutual gain. 
      It returns the training error specified for this node, the gain, the 
      decisions made for the 2 branches for this node and the sub Dataframes 
      for 0 and 1
  '''

  # Get the labels from the dataset
  labels = df_data['Label'].tolist()

  # Split the labels into 2 sub datasets; DataSet for zero labels and another for 1 labels 
  D_0 = GET_labels_by_value(df_data, feature_name, 0)
  D_1 = GET_labels_by_value(df_data, feature_name, 1)

  # Calculate P(feautre = 0) and P(feautre = 1)
  p_feature_0 = len(D_0) / total_length_of_whole_DS
  p_feature_1 = len(D_1) / total_length_of_whole_DS

  # Calculate Entropy: H(Y)
  entropy_y, _ , _ = EntropyAndError_calculation(labels, len(labels), 'Label')

  # Calculate specific conditional entropy: H(Y|feature=0) and H(Y|feature=1)
  SCE_entropy_0, _, decision_0_DS = EntropyAndError_calculation(D_0, len(D_0), 'Zeros sub DataSet')
  SCE_entropy_1, _, decision_1_DS  = EntropyAndError_calculation(D_1, len(D_1), 'Ones sub DataSet')

  # Calculte conditional entropy: H(Y|feature)
  conditional_entropy = p_feature_0 * SCE_entropy_0 + p_feature_1 * SCE_entropy_1

  # Calculte Mutual information gain: I(Y;feature)
  gain = entropy_y - conditional_entropy

  # Get the error for this feature for the 0 and 1 branches to get the trainign error for this branch.
  _, error_on_zero, _ = EntropyAndError_calculation(D_0, total_length_of_whole_DS, 'Zeros sub DataSet')
  _, error_on_one, _  = EntropyAndError_calculation(D_1, total_length_of_whole_DS, 'Ones sub DataSet')

  # Get the subdata of data where a feature value = 0 or 1. The current feature needs to be dropped
  # in order to use the sub datasets in the next evaluation.
  DF_0 = (df_data.loc[df_data[feature_name] == 0]).drop([feature_name], axis = 1)
  DF_1 = (df_data.loc[df_data[feature_name] == 1]).drop([feature_name], axis = 1)

  # Remember the decisions made for this feature for the left and right branches
  decision_dict =  {'left_decision': decision_0_DS, 'right_decision' : decision_1_DS}

  return gain, (error_on_one + error_on_zero), DF_0, DF_1, decision_dict
```

```python
def print_node_details(index, decision, dep):
  global TREE

  for i in range(dep):
    print("|", end = "")
  for node_name in TREE[index]:
    print(node_name, ", train_error: ", TREE[index][node_name]['train_error'], ", decision: ", TREE[index][node_name][decision])

  return node_name

def GetTreeStructure(index, direction, depth):
  global TREE
  if checkKey(TREE[0], 'root'):
    del TREE[0]
    print("\nTree structure: ")

  index -= 1

  # Get node details along with its right decision
  node_name =  print_node_details(index, 'right_decision', depth)

  # Depth increases by 1 as there are possible children for the parent
  depth += 1

  # Go the right and get me the details, son!
  if TREE[index][node_name]['right_feature'] != None:
    GetTreeStructure(TREE[index][node_name]['right_feature_index'], 'right', depth)

  # show the parent name again, it's the beginning of the tree again with a left decision.
  depth -=1
  node_name =  print_node_details(index, 'left_decision', depth)
  depth += 1

  # Now go to the left and get me the details, son!
  if TREE[index][node_name]['left_feature'] != None:
    return  GetTreeStructure(TREE[index][node_name]['left_feature_index'], 'left', depth)

def Get_TrainError(tree):
  sum = 0
  # print("Tree structure: ")
  for node in tree:
    for node_name in node:
      if node[node_name]['left_feature'] == None and node[node_name]['right_feature'] == None:
        sum += node[node_name]['train_error']

  print("Training error for the Decision Tree: ", sum)
  return sum
```

```python
def DecisionTreeTrain(df_data, depth, base_node, base_node_location, direction, total_length_of_whole_DS):
  '''
      This function is used to train a TREE classifier using mutual information as a
      splitting creterion. 
      Inputs:
      df_data:            Input dataframe that contains labels and features to use in the training.
      depth:              The current depth we are in.
      base_node:          The root node for which we are trying to split for deeper trees.
      base_node_location: The location of the current root node in the TREE list.
      direction:          it's either left or right, it's important to know the direction.
                          in order to correctly split the root node.
      total_length:       It's used to correctly calculate certain probabilities. It's just the number
                          of examples in the original dataFrame without any modifications

  '''
  global TREE
  global MAX_DEPTH

  # Make sure you are deep enough
  if depth == MAX_DEPTH:
    return 

  # The dictionary mutualgain contains the following: key -> node name, 
  # value -> list of tuples
  # a tuple contains(Node name,  list(mutual gain, train error for this node, 
  #                 panda Dataframes for Subset zero DataSet and one DataSet, decision dict ))
  mutualgain = {}
  
  # Let's have a default answer to use for this data in case we can't split further using this feature in this recursion frame
  SCE_entropy_0, train_error, decision = EntropyAndError_calculation(df_data['Label'].tolist(), len(df_data['Label'].tolist()), 'label')
  
  # Check for labels in data if unambiguoius, this is the base case: no further splitting
  if train_error == 0:
    # You need to remember the direction from which the current feature is used to split the previous one.
    if direction == 'left':
      # You need to end the feature with a decision
      TREE[base_node_location][base_node]['left_decision'] = decision
      # state that there's no further splitting by using 'None'
      TREE[base_node_location][base_node]['left_feature'] = None
    else: 
      TREE[base_node_location][base_node]['right_decision'] = decision
      TREE[base_node_location][base_node]['right_feature'] = None

    return 

  # Check for empty features. If length = 1, this is just labels and we need to return
  elif  len(df_data.columns) == 1:
    # This is a leaf node and can't split further. Get the decision and leave
    if direction == 'left':
      TREE[base_node_location][base_node]['left_decision'] = decision
      TREE[base_node_location][base_node]['left_feature'] = None
    else: 
      TREE[base_node_location][base_node]['right_decision'] = decision
      TREE[base_node_location][base_node]['right_feature'] = None
    
    return decision

  # Query more features to choose the split feautre
  else:
    # This part is the loop needed to decide which feature to continue with
    for feature in df_data:
      if feature != 'Label':
        mutualgain[feature] = MutualInformationGain_calculator(df_data, feature, decision,  total_length_of_whole_DS)
  
    # Order the dictionary so that the 1st index #0 is the node we split on.
    ordered = {k: v for k, v in sorted(mutualgain.items(), key=lambda item: item[1][0], reverse=True)}

    # Get the parameters calculated, gain, train error, D_0, D_1 and decisions.
    local_mutual_gain = list(ordered.items())[0][1][0]
    local_train_error = list(ordered.items())[0][1][1]
    zeros_data_set    = list(ordered.items())[0][1][2]
    ones_data_set     = list(ordered.items())[0][1][3]
    decision_dict     = list(ordered.items())[0][1][4]
    feature_name      = list(ordered.items())[0][0]

    # Check for the mutual gain if 0 or -ve don't add the current feature name as the next split
    # Just terminate the node with a decision
    if local_mutual_gain <= 0:
      if direction == 'left':
        # Remember to tell the root node that you have a decision and tell the root your name.
        TREE[base_node_location][base_node]['left_decision'] = decision
        TREE[base_node_location][base_node]['left_feature'] = feature_name
        TREE[base_node_location][base_node]['left_feature_index'] = len(TREE)
        decision_dict['left_feature'] = None
        decision_dict['right_feature'] = None        
      else: 
        TREE[base_node_location][base_node]['right_decision'] = decision
        TREE[base_node_location][base_node]['right_feature'] = feature_name
        TREE[base_node_location][base_node]['right_feature_index'] = len(TREE)
        decision_dict['right_feature'] = None
        decision_dict['left_feature'] = None
      
      decision_dict['train_error'] = local_train_error

      TREE.append({feature_name: decision_dict})

      # Remember to update the base node location as you now added a new node
      base_node_location = len(TREE) - 1

      return 
    
    if direction == 'left':
      # Modify the left feature name for the base node
      TREE[base_node_location][base_node]['left_feature'] = feature_name
      TREE[base_node_location][base_node]['left_feature_index'] = len(TREE)
      decision_dict['left_feature'] = None
      decision_dict['right_feature'] = None
    else:
      TREE[base_node_location][base_node]['right_feature'] = feature_name
      TREE[base_node_location][base_node]['right_feature_index'] = len(TREE)
      decision_dict['right_feature'] = None
      decision_dict['left_feature'] = None
  
    decision_dict['train_error'] = local_train_error

    # Append the new node in the TREE list
    TREE.append({feature_name: decision_dict})

    # Update the base node location in order to send it in the recursion
    base_node_location = len(TREE) - 1

    # Once you want to calculate the left and right nodes, you are going further in the deep
    depth += 1

    # Get the splitting features for the left and right branches
    DecisionTreeTrain(zeros_data_set, depth, feature_name, base_node_location, 'left',  total_length_of_whole_DS)
    DecisionTreeTrain(ones_data_set, depth, feature_name, base_node_location, 'right',  total_length_of_whole_DS)

def test_decision_tree(single_train_example, starting_location):
  global TREE
  if checkKey(TREE[0], 'root'):
    del TREE[0]

  starting_location -= 1

  for node_name in TREE[starting_location]:

    # Get node details
    left_feature_decision   = TREE[starting_location][node_name]['left_decision']
    left_feature_name       = TREE[starting_location][node_name]['left_feature']
    right_feature_decision  = TREE[starting_location][node_name]['right_decision']
    right_feature_name      = TREE[starting_location][node_name]['right_feature'] 

    feature_value = single_train_example[node_name]
    
    # Traverse the tree right branch first
    if feature_value == 1:
      # Check to see if there is another feature split
      if checkKey(TREE[starting_location][node_name], 'right_feature_index'):
        right_feature_index = TREE[starting_location][node_name]['right_feature_index']
        return  test_decision_tree(single_train_example, TREE[starting_location][node_name]['right_feature_index'])
      else:
        return right_feature_decision
    
    if feature_value == 0:
      # Check to see if there is another feature split
      if checkKey(TREE[starting_location][node_name], 'left_feature_index'):
        left_feature_index = TREE[starting_location][node_name]['left_feature_index']
        return  test_decision_tree(single_train_example, TREE[starting_location][node_name]['left_feature_index'])
      else:
        return left_feature_decision
```

```python
from sklearn.metrics import accuracy_score
def checkKey(dict, key):
    if key in dict.keys():
      return True
    else:
      return False

def multiple_train_error(df_train, df_test, df_ln_train, df_ln_test, z_te, z_ts):
  global TREE 
  global MAX_DEPTH
  test_error = []
  train_error = []
  depth_list = []

  test_error.append(z_ts)
  train_error.append(z_te)
  depth_list.append(0)

  for i in range(1, 8):
    print("For depth: ", i)
    depth_list.append(i)
    TREE = []
    y_pred = []

    MAX_DEPTH = i
    SET_Node(TREE, 'root', 1, 1, 'x', 'x')
    DecisionTreeTrain(df_train, 0, 'root',  0, 'left', df_ln_train)
    train_error.append(Get_TrainError(TREE))
        
    true_labels = df_test['Label'].tolist()

    for indx in range (df_ln_test):
      y_pred.append(test_decision_tree(df_test.drop(['Label'], axis = 1).iloc[indx], 1))

    test_error.append(1-accuracy_score(true_labels, y_pred))
  
  print("train errors: ", train_error)
  print("test errors: ", test_error)

  return train_error, test_error, depth_list
```
